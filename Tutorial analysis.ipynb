```python
all_months_data = pd.DataFrame()

for file in files:
    df= pd.read_csv(r"C:/Users/zuzan/Downloads/Pandas-Data-Science-Tasks-master/Pandas-Data-Science-Tasks-master/SalesAnalysis/Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data,df])
    
all_months_data
```

```python
#check for NAs
nan_df = all_months_data[all_months_data.isna().any(axis=1)]
nan_df
#Drop NA for all rows
all_months_data=all_months_data.dropna(how='all')
all_months_data
```

```python
all_data=all_months_data[all_months_data['Order Date'].str[0:2]!='Or']
all_data
```

```python
all_data['Month']=all_data["Order Date"].str[0:2]
all_data['Month'] = all_data['Month'].astype('int64') 
all_data
```

```python
all_data['Quantity Ordered']=pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each']=all_data['Price Each'].astype('float')
all_data['Sales']= all_data['Quantity Ordered'] * all_data ['Price Each']
all_data
```

### Month with the highest number of sales

```python
results = all_data.groupby('Month').sum().reset_index()
results
```

```python
import matplotlib.pyplot as plt

plt.bar(results['Month'],results['Sales'])
plt.xticks(results['Month'])
plt.xlabel("Months")
plt.ylabel("Sales in USD ($)")
plt.show()
plt.close()
```

### What city had the highest number of sales

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City']=all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ('+ get_state(x))+ ')'
all_data.head()

```

```python
results = all_data.groupby('City').sum().reset_index()
results
```

```python
import matplotlib.pyplot as plt
#same order for city 
city= [city for city, df in all_data.groupby('City')]
plt.bar(city,results['Sales'])
plt.xticks(city, rotation = 'vertical')
plt.xlabel("City")
plt.ylabel("Sales in USD ($)")
plt.show()
plt.close()
```

### What time should we display advertisements to maximize the likelihood of customer's buying product?

```python
all_data['Order Date']=pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data
```

```python
import matplotlib.pyplot as plt
#same order for city 
hours= [hour for hour, df in all_data.groupby('Hour')]
plt.plot(hours,all_data.groupby('Hour').count())
plt.xticks(hours, rotation = 'vertical')
plt.grid()
plt.xlabel("Hour")
plt.ylabel("Count of orders")
plt.show()
plt.close()
```

### What products are often sold together?

```python
df= all_data[all_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df = df[['Order ID','Grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key, value)
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity_ordered)
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
